```python
%matplotlib inline
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
from scipy import stats
import matplotlib.cm as cm
os.chdir('/Users/user4574/Dropbox (Dunlap-Loros labs)/clocklab')



def cart2pol(x, y):
    rho = np.sqrt(x**2 + y**2)
    phi = np.arctan2(y, x)
    return(rho, phi)

def pol2cart(rho, phi):
    x = rho * np.cos(phi)
    y = rho * np.sin(phi)
    return(x, y)


rna_old = pd.read_csv('JTK.wt_rna_scaled.txt',sep = '\t')
rna_old = rna_old.set_index('rownames(data)')


rna_new = pd.read_csv('JTK.wt_rna_eignorm.txt',sep = '\t')
rna_new = rna_new.set_index('rownames(data)')

psandqs = pd.concat([rna_old['BH.Q'], rna_new['BH.Q'],rna_old['LAG'],rna_new['LAG']], axis=1)
psandqs.columns = ['rna_oldq','rna_newq','rna_oldl','rna_newl']
new = psandqs[psandqs['rna_newq']<.05]
new = new[new['rna_oldq']>.05]
old = psandqs[psandqs['rna_oldq']<.05]
new['set'] = ['new']*len(new)
old['set'] = ['old']*len(old)
oldvnew = old.append(new)

classes = [1 if i < 15 and i > 5 else 0 for i in oldvnew['rna_newl']]
oldvnew['classes']  = classes

forKDE = pol2cart(oldvnew['rna_oldq'].values,oldvnew['rna_newl'].values*np.pi*2/23)

values = np.vstack([forKDE[0], forKDE[1]])

R = 1
r = np.linspace(0,R,100)
theta = np.linspace(-np.pi,np.pi,200)
radius_matrix, theta_matrix = np.meshgrid(r,theta)
positions = np.vstack([radius_matrix.ravel(), theta_matrix.ravel()])

cart_pos = pol2cart(positions[0],positions[1])

cart_pos = np.array([cart_pos[0],cart_pos[1]])

kernel = stats.gaussian_kde(values)
Z = np.reshape(kernel(cart_pos).T, radius_matrix.shape)


```

```python
plt.figure()
ax = plt.axes(polar=True)
xtickslbls = ['0', '2.875','5.75','8.625','11.5','14.375','17.25','20.125']
ax.set_xticklabels(xtickslbls)
ax.set_theta_direction(-1)
ax.set_theta_offset(np.pi/2.0)
ax.set_rlabel_position(70)
CS = plt.contour(theta_matrix, radius_matrix, Z,
                cmap=cm.plasma)
plt.clabel(CS, inline=1, fontsize=10)
plt.title('Old Q-value vs Phase')
```

```python
rna_old = pd.read_csv('JTK.wt_rna_scaled.txt',sep = '\t')
rna_old = rna_old.set_index('rownames(data)')
rna_old.head()

rna_new = pd.read_csv('JTK.wt_rna_eignorm.txt',sep = '\t')
rna_new = rna_new.set_index('rownames(data)')
rna_new.head()

wt = pd.read_csv('JTK.wt_eigenms_imputed.txt',sep = '\t')
wt = wt.set_index('rownames(data)')
wt.head()


csp = pd.read_csv('JTK.csp_eigenms_imputed.txt',sep = '\t')
csp = csp.set_index('rownames(data)')
csp.head()
psandqs = pd.concat([rna_old['BH.Q'], rna_new['BH.Q'], wt['BH.Q'],csp['BH.Q'],rna_old['LAG'],rna_new['LAG'], wt['LAG'],csp['LAG']], join='inner', axis=1)
psandqs.columns = ['rna_oldq','rna_newq','wtq','cspq','rna_oldl','rna_newl','wtl','cspl']
```

```python
wandc = psandqs[psandqs['wtq']<.05]
wandc = wandc[wandc['cspq']<.05]
len(wandc)
```

```python
cspl = wandc['cspl'].values.tolist()
wtl = wandc['wtl'].values.tolist()
```

```python
pdelt = [min([np.abs(i%23-j%23),np.abs((i%23+23)-j%23),np.abs(i%23-(j%23+23))]) for i,j in zip(cspl,wtl)]
```

```python
len(pdelt)
```

```python
wandc['pdelt'] = pdelt
```

```python
forKDE = pol2cart(wandc['wtl'].values,wandc['pdelt'].values*np.pi*2/23)

values = np.vstack([forKDE[0], forKDE[1]])

R = 23
r = np.linspace(0,R,100)
theta = np.linspace(-np.pi,np.pi,200)
radius_matrix, theta_matrix = np.meshgrid(r,theta)
positions = np.vstack([radius_matrix.ravel(), theta_matrix.ravel()])

cart_pos = pol2cart(positions[0],positions[1])

cart_pos = np.array([cart_pos[0],cart_pos[1]])

kernel = stats.gaussian_kde(values)
Z = np.reshape(kernel(cart_pos).T, radius_matrix.shape)
```

```python
plt.figure()
ax = plt.axes(polar=True)
xtickslbls = ['0', '2.875','5.75','8.625','11.5','14.375','17.25','20.125']
ax.set_xticklabels(xtickslbls)
ax.set_theta_direction(-1)
ax.set_theta_offset(np.pi/2.0)
ax.set_rlabel_position(70)
CS = plt.contour(theta_matrix, radius_matrix, Z,
                cmap=cm.plasma)
plt.clabel(CS, inline=1, fontsize=10)
plt.title('Phase vs Pdelt')
```
