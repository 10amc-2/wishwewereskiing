```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

sns.set()

# Generate an example radial datast
r = np.linspace(0, 10, num=100)
df = pd.DataFrame({'r': r, 'slow': r, 'medium': 2 * r, 'fast': 4 * r})

# Convert the dataframe to long-form or "tidy" format
df = pd.melt(df, id_vars=['r'], var_name='speed', value_name='theta')

# Set up a grid of axes with a polar projection
g = sns.FacetGrid(df, col="speed", hue="speed",
                  subplot_kws=dict(projection='polar'), size=4.5,
                  sharex=False, sharey=False, despine=False)

# Draw a scatterplot onto each axes in the grid
g.map(plt.scatter, "theta", "r")
```

```python
rna_old = pd.read_csv('JTK.wt_rna_scaled.txt',sep = '\t')
rna_old = rna_old.set_index('rownames(data)')

rna_new = pd.read_csv('JTK.wt_rna_eignorm.txt',sep = '\t')
rna_new = rna_new.set_index('rownames(data)')
```

```python
psandqs = pd.concat([rna_old['BH.Q'], rna_new['BH.Q'],rna_old['LAG'],rna_new['LAG']], axis=1)
psandqs.columns = ['rna_oldq','rna_newq','rna_oldl','rna_newl']
```

```python
new = psandqs[psandqs['rna_newq']<.05]
new = new[new['rna_oldq']>.05]
old = psandqs[psandqs['rna_oldq']<.05]
```

```python
new['set'] = ['new']*len(new)
old['set'] = ['old']*len(old)
```

```python

oldvnew = old.append(new)

classes = [1 if i < 15 and i > 5 else 0 for i in oldvnew['rna_newl']]
oldvnew['classes']  = classes
```

```python
# Set up a grid of axes with a polar projection
g = sns.FacetGrid(oldvnew[['set','rna_newl','rna_oldq']], col="set", hue="set",subplot_kws=dict(projection='polar'), size=4.5, sharex=False, sharey=False, despine=False)

# Draw a scatterplot onto each axes in the grid
g.map(plt.scatter, "rna_newl", "rna_oldq",alpha=.2)
```
